```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
```

```python
dataset['sex_male'] = dataset['sex_male'].astype(int)
dataset['smoker_yes'] = dataset['smoker_yes'].astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
dependent = dataset[['charges']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size = 0.3, random_state = 0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
regressor1 = DecisionTreeRegressor()
regressor1.fit(x_train,y_train)
regressor2 = AdaBoostRegressor()
regressor2.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor1)
plt.show()
```

```python
y_predict1 = regressor1.predict(x_test)
from sklearn.metrics import r2_score
r_score1 = r2_score(y_test,y_predict1)
r_score1
```

```python
y_predict2 = regressor2.predict(x_test)
from sklearn.metrics import r2_score
r_score2 = r2_score(y_test,y_predict2)
r_score2
```
